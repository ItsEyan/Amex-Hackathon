```python
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import pickle
```

```python
eval_df = pd.read_csv('65d4b8b2ebfe9_amex_campus_challenge_eval_round1_2.csv')
```

```python
eval_df.loc[eval_df['customer'] == 467915]
```

```python
eval_df = eval_df.fillna(eval_df.mean())
```

```python
eval_df["customer"].nunique()
```

```python
with open('models.pkl', 'rb') as f:
    logistic_models = pickle.load(f)
```

```python
with open('cluster_model.pkl', 'rb') as f:
    kmeans = pickle.load(f)
```

```python
with open('customer_filter.pkl', 'rb') as f:
    filtered = pickle.load(f)
```

```python
df_averaged = eval_df[[col for col in eval_df.columns if col.startswith('customer') or col.startswith('distance_')]].groupby(['customer']).mean()
df_averaged
```

```python
df_filtered = df_averaged[filtered]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_filtered)
```

```python
cluster_labels = kmeans.fit_predict(scaled_features)
df_filtered['Cluster_Label'] = cluster_labels
```

```python
df_filtered
```

```python
eval_df['ind_recommended'] = 0
```

```python
for cluster in cluster_labels.unique():
    for customer in df_filtered[df_filtered['Cluster_Label'] == cluster].index:
        df_customer = eval_df[eval_df['customer'] == customer]
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(df_customer)
        customer_predicted = logistic_models[cluster].predict_proba(scaled_features)
    
    
```
