```python
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import pickle
```

```python
eval_df = pd.read_csv('65d4b8b2ebfe9_amex_campus_challenge_eval_round1_2.csv')
```

```python
eval_df.loc[eval_df['customer'] == 467915]
```

```python
eval_df = eval_df.fillna(eval_df.mean())
```

```python
eval_df["customer"].nunique()
```

```python
with open('models.pkl', 'rb') as f:
    logistic_models = pickle.load(f)
```

```python
with open('cluster_model.pkl', 'rb') as f:
    kmeans = pickle.load(f)
```

```python
with open('customer_filter.pkl', 'rb') as f:
    filtered = pickle.load(f)
```

```python
df_averaged = eval_df[[col for col in eval_df.columns if col.startswith('customer') or col.startswith('distance_')]].groupby(['customer']).mean()
df_averaged
```

```python
df_filtered = df_averaged[filtered]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_filtered)
```

```python
cluster_labels = kmeans.fit_predict(scaled_features)
df_filtered['Cluster_Label'] = cluster_labels
```

```python
df_filtered
```

```python
def assign_range(value):
    if value < 150:
        return 1
    elif 175 <= value <= 225:
        return 2
    elif 250 <= value <= 280:
        return 3
    elif 285 <= value <= 350:
        return 4
    elif 375 <= value <= 450:
        return 5
    elif value > 450:
        return 6
```

```python
eval_df['merchant_profile_01'] = eval_df['merchant_profile_01'].apply(assign_range)
```

```python
eval_df['ind_recommended'] = 0
```

```python
clustered_df =[]
for cluster in sorted(pd.unique(cluster_labels)):
    cur = df_filtered[df_filtered['Cluster_Label'] == cluster]
    clustered_df.append(eval_df[eval_df['customer'].isin(cur.index)])
```

```python
cols = [col for col in eval_df.columns if col.startswith('customer') or col.startswith('distance_')]
for df in clustered_df:
    df.drop(columns=cols, inplace=True)
```

```python
clustered_df[0]
```

```python
test_cols = ['ind_recommended', 'merchant_spend_02', 'merchant_spend_03', 'merchant_spend_06', 'merchant_spend_10', 'merchant_profile_01', 'merchant']
```

```python
for index, cluster in enumerate(clustered_df):
    scaler = MinMaxScaler()
    cur = cluster[test_cols]
    scaled_features = scaler.fit_transform(cur.drop(columns=['merchant']))
    customer_predicted = logistic_models[index].predict_proba(scaled_features)
    cluster['not_recommended_prob'] = customer_predicted[:, 1]
```

```python
for index, cluster in enumerate(clustered_df):
    scaler = StandardScaler()
    cluster['ind_recommended'] = 1
    cur = cluster[test_cols]
    scaled_features = scaler.fit_transform(cur.drop(columns=['merchant']))
    customer_predicted = logistic_models[index].predict_proba(scaled_features)
    cluster['recommended_prob'] = customer_predicted[:, 1]
```

```python
for cluster in clustered_df:
    cluster['incremental_activation'] = cluster['recommended_prob'] - cluster['not_recommended_prob']
```
